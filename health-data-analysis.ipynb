```python
#taking the values for men
male_change =df.loc[df['sex']==1,'weight_change']
#taking the values for women
female_change =df.loc[df['sex']==2,'weight_change']
#executing the test
outcome = stats.ttest_ind(male_change,female_change)
#printing values
print('t=',outcome.statistic)
print('p=',outcome.pvalue)
```

For this task, I used loc function to select data for both males and females. By doing this, I got the details.Then I used the stats.ttest_ind method on these.The t value showed the direction and the difference whereas p value showed the accuracy.In this case, t value was -3.79 and p value was 0.000145 so it means that males had less weight change in comparing to females and the difference is real as the p value was smaller.

```python
#using the seed
np.random.seed(seed)
random.seed(seed)
#creating a list of row poditions
row=list(range(len(df)))
#shuffling it
random.shuffle(row)
#to make it into two parts , finding a middle point
half_mark =len(df)//2
#first half after shuffling
first_half =df.iloc[row[:half_mark]]['weight_change']
#second half after shuffling
second_half =df.iloc[row[half_mark:]]['weight_change']
#executing the task
final_result=stats.ttest_ind(first_half,second_half)
#printing result
print('t=',final_result.statistic)
print('p=',final_result.pvalue)
```

For this task, first I set up the random seed using my id so the result remains reproducible. I made a list where I collected all the row numbers.Then with the help of shuffle method I shuffled.Following that I splitted the list.So there's two part. The first half and the second half.Then I used the iloc method to get those details.I operated the t test and printed the result.The result showed that the t value is -0.59 which means the differences between those half were very less and the negative symbol indicated that the weight change of first half was less than the weight change of second one.The p value was 0.55 so there's no significant difference.

```python
#creating an empty list to store values
p_values=[]
#running the test for 1000 times
for i in range(1000):
  #creating list of row values
  list_row = list(range(len(df)))
  #shuffling them
  random.shuffle(list_row)
  #getting the middle point to split it
  half_point =len(df)//2
  #first half data
  first_half_data =df.iloc[list_row[:half_point]]['weight_change']
  #second half data
  second_half_data =df.iloc[list_row[half_point:]]['weight_change']
  #running the final test
  result =stats.ttest_ind(first_half_data,second_half_data)
  #only the p value is taken
  p=result.pvalue
  if p>0:
    #appending into the list
    p_values.append(p)
#an empty list to store values
neg_values=[]
#looping to append the values
for i in p_values:
  neg_values.append(-math.log10(i))
```

In this task, first I created an empty list to store values. I ran the test for 1000 times according to the instructions.Inside the loop I did same like before I did in task 3.2. I operated the t test again. From the result I took only p values.Then I checked the p value was greater than zero or not.Because log 10(0) provides undefined which would crash the code. .Then I stored those values. I converted every value in the p values list into negative log10 to get a proper histogram.

```python
#creating a new figure
plt.figure()
#creating a hisogram
plt.hist(neg_values,bins=30)
#title with student id
plt.title('-log10(p) from 1000 random splits (SID'+student_id+')')
#labelling
plt.xlabel('-log10(p)')
plt.ylabel('Frequency')
#displaying the result
plt.show()
```
